## Statistical Analysis

```python
path = r'/home/riya/SAAF_new/python_template/results/results_24_hours_experiment_5_23__oregon_128MBvs10GB'
all_files = glob.glob(os.path.join(path, "*.csv"))

df = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)
df
```

```python
df['time_run']= pd.to_datetime(df['time_run'], format='%m/%d/%Y %H:%M:%S')
```

```python
count_by_hour_df = df[['cpuType', 'memory','time_run']].value_counts().reset_index(name='count')
count_by_hour_df
```

```python
mem_128_df = count_by_hour_df.loc[count_by_hour_df['memory'] == '10GB']
mem_128_df['%'] = 100 * mem_128_df['count'] / mem_128_df.groupby('time_run')['count'].transform('sum')

mem_128_df
```

```python
mem_128_df_pivot = mem_128_df.pivot(index='time_run', columns=['cpuType'], values=['%'])

mem_128_df_pivot
```

```python
mem_128_df_pivot = mem_128_df_pivot * 0.01
mem_128_df_pivot
```

```python

mem_128_df_pivot.to_csv('pivot.csv', index=True) 
```
